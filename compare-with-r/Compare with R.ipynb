```python
import sys
sys.path.insert(0, '../../')
from gsea.gsea import gsea
from file.file import gct, gmt
from plot.plot import plot

%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
gene_x_sample = gct.read_gct('./kras_isogenic_vs_imortalized.gct')

print(gene_x_sample.shape)
gene_x_sample
```

```python
gene_sets = gmt.read_gmts('./c2.all.v5.1.symbols.gmt')

print(gene_sets.shape)
gene_sets = gene_sets.iloc[:5]
gene_sets
```

```python
gene_set_x_sample_pablo = gct.read_gct(
    './kras_isogenic_vs_imortalized.ssGSEA.test.gct')
gene_set_x_sample_pablo = gene_set_x_sample_pablo.ix[gene_sets.index, :]

gene_set_x_sample = gsea.single_sample_gsea(
    gene_x_sample, gene_sets, normalization='rank', power=0.75, statistic='AUC')
```

```python
plot.plot_heatmap(
    gene_set_x_sample_pablo,
    annot=True)
plot.plot_heatmap(
    gene_set_x_sample,
    annot=True)

plot.plot_heatmap(
    gene_set_x_sample_pablo,
    annot=True,
    normalization_method='-0-',
    normalization_axis=0)
plot.plot_heatmap(
    gene_set_x_sample,
    annot=True,
    normalization_method='-0-',
    normalization_axis=0)

plot.plot_heatmap(
    gene_set_x_sample_pablo,
    annot=True,
    normalization_method='-0-',
    normalization_axis=1)
plot.plot_heatmap(
    gene_set_x_sample,
    annot=True,
    normalization_method='-0-',
    normalization_axis=1)
```
